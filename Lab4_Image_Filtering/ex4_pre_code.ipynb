```python
from skimage.util import img_as_float32
import matplotlib.pyplot as plt
import numpy as np
from numpy import fft
from numpy import random

W = 64        # window squared dimenstion
M = 256       # image squared dimenstion


img = np.zeros((M,M), int)

for i in range(M):
  for j in range(M):
    if ( (i >= ((M-W)/2)) & (i <= ((M+W)/2)) & (j >= ((M-W)/2)) & (j <= ((M+W)/2)) ):
      img[i,j] = 255
    else:
      img[i,j] = 0

plt.imshow(img, cmap='gray')
```

```python

ft = fft.fft2(img)
log_mag_ft = 20*np.log(np.abs(ft)+1) # add 1 - in order to prevent log(0)
ft_shift = fft.fftshift(ft)
log_mag_ft_shift = 20*np.log(np.abs(ft_shift)+1)

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 10))
ax = ax.ravel()

ax[0].imshow(img, cmap='gray')
ax[0].set_title('The original image'), ax[0].axis('on')
ax[1].imshow(log_mag_ft, cmap='gray')
ax[1].set_title('|DFT(img)| [dB]')
ax[2].imshow(log_mag_ft_shift, cmap='gray')
ax[2].set_title('|DFT(img)| [dB] + fftshift')
plt.tight_layout()
plt.show()
```

```python
M = 256       # image squared dimenstion

R = 32

img = np.zeros((M,M), int)

for i in range(M):
  for j in range(M):
    r = np.sqrt((i-128)**2 + (j-128)**2)

    if ( (r <= R)):
      img[i,j] = 255
    else:
      img[i,j] = 0

plt.imshow(img, cmap='gray')
```

```python

ft = fft.fft2(img)
log_mag_ft = 20*np.log(np.abs(ft)+1) # add 1 - in order to prevent log(0)
ft_shift = fft.fftshift(ft)
log_mag_ft_shift = 20*np.log(np.abs(ft_shift)+1)

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 10))
ax = ax.ravel()

ax[0].imshow(img, cmap='gray')
ax[0].set_title('The original image'), ax[0].axis('on')
ax[1].imshow(log_mag_ft, cmap='gray')
ax[1].set_title('|DFT(img)| [dB]')
ax[2].imshow(log_mag_ft_shift, cmap='gray')
ax[2].set_title('|DFT(img)| [dB] + fftshift')
plt.tight_layout()
plt.show()
```
